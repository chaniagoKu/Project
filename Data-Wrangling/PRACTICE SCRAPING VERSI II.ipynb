```python
iD =[]
nm =[]  
fm =[]
pnm =[]
ref =[]
st =[]
ak =[]
for i in range (400) :
    url = "http://herbaldb.farmasi.ui.ac.id/v3/index.php?v=spesies"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    tabel = soup.find_all('table', border = '0', width = '100%')[1]
    for j in tabel.find_all('tr') :
        idi = j.find_all('td', style="text-align:center:")
        nama = j.find_all('td',  style="text-align:left;padding-left:10px;")[0].text
        family = j.find_all('td', style="text-align:left;padding-left:10px;")[2].text
        penemu = j.find_all('td', style="text-align:left;padding-left:10px;")[3].text
        referensi = j.find_all('td', style="text-align:left;padding-left:10px;")[4].text
        status = j.find_all('td', style="text-align:left;padding-left:10px;")[5].text
        aksi = j.find_all('td', style="text-align:left;padding-left:10px;")[6].text
        iD.append(idi)
        nm.append(nama)
        fm.append(family)
        pnm.append(penemu)
        ref.append(referensi)
        st.append(status)
        ak.append(aksi)
    print("Load page ke -", i)
```

```python
df = pd.DataFrame(np.array([iD,nm,fm,pnm,ref,st,ak]), index = ['ID Spesies','Spesies','Family','Penemu','Referensi','Status','Aksi']).T
print("Berikut merupakan daftar tabel spesies dari tanaman obat di Indonesia")
df
```

```python
df.to_csv("simpan_latihan_SCRAPING.csv")
```
