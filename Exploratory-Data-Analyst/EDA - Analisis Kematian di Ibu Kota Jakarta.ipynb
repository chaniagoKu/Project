#### Grafik Visualisasi


1. Kolom Tahun dan Bulan tidak perlu di visualisasikan

-

2. Kolom kota_Kabupaten

```python
a = pd.pivot_table(df,index = Kota_Kabupaten,values = 'kota_kabupaten', aggfunc = 'count', observed = True)
a
```

```python
fig = plt.figure(figsize=(10,10))
print(a)

ax = plt.axes()
Warna = ["orange", "yellow","magenta" , "turquoise","Grey","springgreen"]
ax.pie(a['kota_kabupaten'], labels = a.index, autopct='%1.0f%%',colors = Warna, explode = [1,0,0,0.2,0.2,0],pctdistance=0.6,shadow = True)
centre_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax.set_title('Jumlah laporan yang masuk per-Kota/Kabupaten yang akan diolah')
plt.show()
```

2. Kolom kecamatan

```python
pd.pivot_table(df,index = [Kota_Kabupaten,Kecamatan],values = 'kecamatan', aggfunc = 'count', observed = True)
```

```python
b = df.groupby(['kecamatan'])['kecamatan'].count().to_frame().rename(columns={'kecamatan':'count'}).reset_index()
b
```

```python
# Memvisualisasikan dengan barplot di library Seaborn.
plt.figure(figsize=(10,8))
plt.grid(axis = 'y', linewidth=0.5, color='black')
ax = sns.barplot(data = b,x = 'kecamatan',y = 'count')
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```

3. Kolom Kelurahan

```python
c = df.groupby(['kelurahan'])['kelurahan'].count().to_frame().rename(columns={'kelurahan':'count'}).reset_index()
# plt.setp(ax.get_xticklabels(), rotation=90)
plt.yticks(rotation=45)
plt.bar(c['kelurahan'],c['count'])
```

```python
# Memvisualisasikan dengan barplot di library Seaborn.
plt.figure(figsize=(10,8))
plt.grid(axis = 'y', linewidth=0.5, color='black')
ax = sns.barplot(data = c,x = 'kelurahan',y = 'count')
plt.setp(ax.get_xticklabels(), rotation=45)
plt.show()
```

4. Kolom jenis_kelamin

```python
d = df.groupby(['jenis_kelamin'])['jenis_kelamin'].count().to_frame().rename(columns={'jenis_kelamin':'count'}).reset_index()
d
```

```python
Jenis_Kelamin.value_counts()
```

```python
fig = plt.figure(figsize=(10,10))
print(d)

ax = plt.axes()
Warna = ["magenta","springgreen"]
ax.pie(d['count'], labels = d['jenis_kelamin'], autopct='%1.0f%%',colors = Warna, explode = [0,0],pctdistance=0.6)
centre_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax.set_title('Jumlah laporan yang masuk per-Jenis Kelamin Pelapornya')
plt.show()
```

5. Kolom Jumlah berdasarkan Jenis Kelamin

```python
e = df.groupby(['jenis_kelamin'])['jumlah'].sum().to_frame().rename(columns={'jenis_kelamin':'count'}).reset_index()
e
```

```python
fig = plt.figure(figsize=(10,10))
print(e)

ax = plt.axes()
Warna = ["magenta","springgreen"]
ax.pie(e['jumlah'], labels = e['jenis_kelamin'], autopct='%1.0f%%',colors = Warna, explode = [0,0],pctdistance=0.6)
centre_circle = plt.Circle((0, 0), 0.68, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax.set_title('Jumlah Meninggal Berdasarkan Jenis Kelamin')
plt.show()
```

6. Kolom Jumlah berdasarkan Kecamatan

```python
f = df.groupby(['kota_kabupaten'])['jumlah'].sum().to_frame().rename(columns={'jenis_kelamin':'count'}).reset_index()
f
```

```python
# Memvisualisasikan dengan barplot di library Seaborn.
plt.figure(figsize=(10,8))
plt.grid(axis = 'y', linewidth=0.5, color='black')
ax = sns.barplot(data = f,x = 'kota_kabupaten',y = 'jumlah')
plt.setp(ax.get_xticklabels(), rotation=45)
plt.show()
```
