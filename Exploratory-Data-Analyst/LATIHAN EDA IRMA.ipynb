```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('FILE_LATIHAN/GATAU BILANG APA.xlsx')
df
```

```python
# Mencari rata-rata tinggi badan berdasarkan jenis kelamin
tinggi = df.groupby('Jenis Kelamin')['Tinggi Badan (cm) (hanya angka)'].mean()
tinggi
```

```python
# Mencari rata-rata berat badan berdasarkan jenis kelamin
berat = df.groupby('Jenis Kelamin')['Berat Badan (kg) (hanya angka)'].mean()
berat
```

```python
# Membuat kolom baru dengan nama Tinggi Badan (Ubah menjadi satuan m)
df['Tinggi Badan'] = df['Tinggi Badan (cm) (hanya angka)']/10
df
```

```python
df['IBM'] = df['Berat Badan (kg) (hanya angka)']/(df['Tinggi Badan'])**2
df
```

```python
# Mencari rata-rata IBM berdasarkan jenis kelamin
ibm = df.groupby('Jenis Kelamin')['IBM'].mean()
ibm
```

```python
df['IBM']=df['IBM']*1000
df
```

```python
# Mencari hubungan yang signifikan
df_kor = df.corr()
df_kor
```

```python
import seaborn as sns
sns.heatmap(df_kor, annot=True)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FILE_LATIHAN/electronic-card-transactions-october-2022-csv-tables.csv')
data
```

```python
# Cek missing value
data.isnull().sum()
```

```python
mis_val = data.isnull().sum()
mis_val.to_frame()

# Percentage of missing values
mis_val_percent = 100 * data.isnull().sum() / len(data)
mis_val_percent.to_frame()

# Make a table with the results
mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
mis_val_table

# Rename the columns
mis_val_table_ren_columns = mis_val_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values'})
mis_val_table_ren_columns
```
