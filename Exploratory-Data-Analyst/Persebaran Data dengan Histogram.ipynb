```python
import numpy as np
import random
from matplotlib import pyplot as plt
np.random.seed(7)
Sample = np.random.uniform(15, 80, 10000)
```

```python
def extractDigits(lst):
    return [[el] for el in lst]
```

```python
def midrange(lst):
    lst.sort()
    Tertinggi = max(lst)
    Terendah = min(lst)
    return (Tertinggi + Terendah)/2
```

```python
# def Display(lst):
#     for i in range (5) :
#         print(lst[i])
def Display(lst):
    Jumlah = int(input("Masukkan Banyak Item Yang Ingin Ditampilkan : "))
    temp = []
    for i in range (Jumlah) :
        temp.append(lst[i])
    print(temp)
```

```python
Coba = []
for _ in range(500):
    x = np.random.choice(Sample, size=1000, replace=True)
    Coba.append(x)
```

```python
my_samples_Mean = []
my_samples_MidRange = []
my_samples_Median = []
for i in range (len(Coba)-1):
    my_samples_Mean.append(np.mean(Coba[i]))
    my_samples_Median.append(np.median(Coba[i]))
    my_samples_MidRange.append(midrange(Coba[i]))
```

```python
# Menara atau batang histogram disebut bin. Ketinggian setiap nampan menunjukkan berapa banyak nilai dari data itu yang termasuk dalam rentang itu. Lebar setiap bin adalah = (nilai maks data  nilai min data) / jumlah total bin. Nilai default dari jumlah bin yang akan dibuat dalam histogram adalah 10.
```

```python
# HISTOGTAM
plt.figure(figsize=(15,7))
colors = ['#FF2E63']
plt.hist(my_samples_Mean, histtype ='bar',edgecolor="Black", color = colors, label = 'Mean', rwidth=1)
# plt.grid(axis='y', alpha=0.75)
plt.grid(alpha=0.75)
plt.legend(prop ={'size': 10})
plt.title('Histogram - Mean', fontweight ="bold")
plt.show()

# LINE CHART
plt.figure(figsize=(15,7))
sns.lineplot(data=my_samples_Mean,color = '#FF2E63', label = 'Mean')
```

```python
# HISTOGTAM
plt.figure(figsize=(15,7))
colors = ['#08D9D6']
plt.hist(my_samples_Median, histtype ='bar',edgecolor="Black", color = colors, label = 'Median', rwidth=1)
plt.grid(alpha=0.75)
plt.legend(prop ={'size': 10})
plt.title('Histogram - Median', fontweight ="bold")
plt.show()

# LINE CHART
plt.figure(figsize=(15,7))
sns.lineplot(data=my_samples_Median,color = '#08D9D6',dashes = True, label = 'Median')
```

```python
# HISTOGTAM
plt.figure(figsize=(15,7))
colors = ['#C06C84']
plt.hist(my_samples_MidRange, histtype ='bar',edgecolor="Black", color = colors, label = 'MidRange', rwidth=1)
plt.grid(alpha=0.75)
plt.legend(prop ={'size': 10})
plt.title('Histogram - MidRange', fontweight ="bold")
plt.show()

# LINE CHART
plt.figure(figsize=(15,7))
sns.lineplot(data=my_samples_MidRange,color = '#C06C84',dashes = True, label = 'MidRange')
```
