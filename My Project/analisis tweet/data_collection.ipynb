```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Display the DataFrame
display(df.head())
```

```python
# Gunakan fungsi to_csv() dari pandas untuk menyimpan DataFrame ke dalam file CSV
df.to_csv('pppk_guru.csv', index=False)  # index=False agar indeks tidak disimpan dalam file CSV
```
